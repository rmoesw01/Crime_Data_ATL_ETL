```python
#Read merged CSVs to a dataframe using glob
#  This code was inspired by the stackoverflow thread:    
#      https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe

apd_raw_df = pd.concat(map(pd.read_csv, glob.glob('Data_files/APD_data_2018_to_2020/*.csv')))
```

```python
merged_file_1 = "Data_Files/2015-2017-APD.csv"
merged_1_df = pd.read_csv(merged_file_1)
merged_1_df = merged_1_df.rename(columns={'Crime Type': 'crime_type', 'Report Number': 'report_number', 'Report Date': 'report_date',\
                                         'Location': 'location', 'Beat': 'beat', 'NPU': 'npu', 'Lat': 'lat','Long': 'long','Neighborhood':'neighborhood'})
merged_1_df.head()
```

```python
merged_file_2 = "Data_Files/merged_2018_to_2020.csv"
merged_2_df = pd.read_csv(merged_file_2)
merged_2_df = merged_2_df.rename(columns={'Crime Type': 'crime_type', 'Report Number': 'report_number', 'Report Date': 'report_date',\
                                         'Location': 'location', 'Beat': 'beat', 'NPU': 'npu', 'Lat': 'lat','Long': 'long', 'Neighborhood':'neighborhood'})
merged_2_df.head()
```

```python
shooting_file = "Data_Files/police_shootings_us.csv"
shooting_df = pd.read_csv(shooting_file)

shooting_df = shooting_df.rename(columns={'body_camera': 'body_cam'})
shooting_df
```

```python
shooting_df.to_sql(name='apd_police_shooting', con=engine, if_exists='append', index=False)
```

```python
merged_1_df.to_sql(name='apd_crime_data', con=engine, if_exists='append', index=False)
```

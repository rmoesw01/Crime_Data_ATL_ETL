```python
import os
import pandas as pd
```

```python
date_range = ('2015-01', '2017-12')
start_date = date_range[0].split('-')
end_date = date_range[1].split('-')
```

```python
start_date
```

```python
APD_df = pd.DataFrame(columns=['UC2_Literal','Report Number','Report Date','Location','Beat','Neigborhood','NPU','Lat','Long','junk'])
APD_df
```

```python
for year in range(int(start_date[0]),int(end_date[0])+1):
    for month in range(1,13):
        pad_month = str(month).rjust(2, '0')
        input_file = os.path.join('Data_Files', 'Monthly', f'{year}-{pad_month}-APD.csv')
        current_df = pd.read_csv(input_file)
        APD_df = APD_df.append(current_df)
APD_df
```

```python
APD_df.count()
```

```python
APD_df.drop(columns=['junk','Unnamed: 9'], inplace=True)
APD_df.rename(columns={'Neigborhood':'Neighborhood'}, inplace=True)
APD_df.count()
```

```python
APD_no_nas_df = APD_df.fillna('unknown')
APD_no_nas_df.count()
```

```python
APD_no_nas_df
```

```python
APD_no_nas_df.loc[APD_no_nas_df['Neighborhood']=='unknown']
```

```python
output_file = os.path.join('Data_Files','2015-2017-APD.csv')
APD_no_nas_df.to_csv(output_file, index=False)
```

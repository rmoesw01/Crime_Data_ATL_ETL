```python
# Dependencies
import os
import pandas as pd
```

```python
# Set path for US Police Shootings Data File
input_file = os.path.join('Data_Files', 'police_shootings_us.csv')
```

```python
# Setting chosen state and city
state = 'GA'
city = 'Atlanta'
```

```python
# Pull csv into a pandas dataframe
shootings = pd.read_csv(input_file)
shootings
```

```python
shootings['age'] = round(shootings['age'],0)
```

```python
shootings.count()
```

```python
# Filter the shootings dataframe to just show the records for the selected state
shootings_state = shootings.loc[shootings['state']==state]
shootings_state
```

```python
# Filter the shootings dataframe to just show the records for the selected city
shootings_city = shootings_state.loc[shootings_state['city']==city]
shootings_city
```

```python
# Create output file paths
output_state = os.path.join('Data_Files', 'Shootings_Filtered', f'police_shootings_{state}.csv')
output_city = os.path.join('Data_Files', 'Shootings_Filtered', f'police_shootings_{state}_{city}.csv')
```

```python
# Output DataFrames to csv files
shootings_state.to_csv(output_state, index=False)
shootings_city.to_csv(output_city, index=False)
```
